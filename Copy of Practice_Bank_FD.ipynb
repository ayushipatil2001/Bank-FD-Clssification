# Normalization

 STandard Scalar

- Split in X & Y

```python
X_train = train_new.drop("y_yes",axis=1)
Y_train = train_new["y_yes"]
```

```python
X_test = test_new.drop("y_yes",axis=1)
Y_test = test_new["y_yes"]
```

## Normalization of target column should not be done!!!

```python
from sklearn import preprocessing
x = X_train.values #returns a numpy array
std_scaler = preprocessing.StandardScaler()
std_scaler.fit(x)
x_scaled_std = std_scaler.transform(x)
X_train_std = pd.DataFrame(x_scaled_std,columns=X_train.columns)
```

```python
x = X_test.values
x_scaled_std = std_scaler.transform(x)
X_test_std = pd.DataFrame(x_scaled_std,columns=X_test.columns)
```

```python
X_train_std.describe()
```

# Apply Classification algorithm

# Naive Bayes

there are 5 variants

- Gaussian NB :: for numeric columns

- Multinomial NB for multi nominal data / text data

- Complement NB improved multi nominal

- Bernoulli NB for binary columns

- Categorical NB for all columns which are categorical

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, \
                           ComplementNB, BernoulliNB, \
                           CategoricalNB
```

```python
clf_gnb = GaussianNB()
clf_gnb.fit(X_train_std, Y_train) # used for training
Y_pred_gnb = clf_gnb.predict(X_test_std)
```

```python
clf_bnb = BernoulliNB()
clf_bnb.fit(X_train_std, Y_train) # used for training
Y_pred_bnb = clf_bnb.predict(X_test_std)
```

```python
clf_mnb= MultinomialNB()
clf_mnb.fit(X_train_std, Y_train) # used for training
Y_pred_mnb = clf_mnb.predict(X_test_std)
```

```python
clf_cnb = ComplementNB()
clf_cnb.fit(X_train_std, Y_train) # used for training
Y_pred_cnb = clf_cnb.predict(X_test_std)
```

```python

clf_catnb = CategoricalNB()
clf_catnb.fit(X_train_std, Y_train) # used for training
Y_pred_catnb = clf_catnb.predict(X_test_std)
```

# Support Vector Machine Classifier

```python
from sklearn.svm import SVC
```

```python
clf_svc = SVC()
clf_svc.fit(X_train_std, Y_train) # used for training
Y_pred_svc = clf_svc.predict(X_test_std)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf_dt = DecisionTreeClassifier()
clf_dt.fit(X_train_std, Y_train) 
Y_pred_dt = clf_dt.predict(X_test_std)
```

# Random forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf_rf = RandomForestClassifier()
clf_rf.fit(X_train_std, Y_train) 
Y_pred_rf = clf_rf.predict(X_test_std)
```

# Compare the classifiers

```python
def evaluate_clf(Y_test, Y_pred):
  from sklearn.metrics import precision_recall_fscore_support
  p,r,f,s = precision_recall_fscore_support(Y_test,Y_pred)
  return p,r,f,s
```

```python
print("Gaussian NB")
print(evaluate_clf(Y_test, Y_pred_gnb))
```

```python
print("Burnoilli NB")
print(evaluate_clf(Y_test, Y_pred_bnb))
```

```python
print("SVM classifier")
print(evaluate_clf(Y_test, Y_pred_svc))
```

```python
print("DT classifier")
print(evaluate_clf(Y_test, Y_pred_dt))
```

```python
print("RF classifier")
print(evaluate_clf(Y_test, Y_pred_rf))
```
